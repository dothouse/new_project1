```python
df3.rename(columns={'행정구': '시군구', '행정동': '읍면동'}, inplace=True)
df3
```

```python
filtered_data = df3[df3['읍면동'].str.contains('\.') | df3['읍면동'].str.contains('·')]
```

```python
df3 = df3.groupby(['시군구', '읍면동']).size().reset_index(name='대학교 개수')
df3
```

```python
import re
import geopandas as gpd
from fiona.crs import from_epsg
import geopandas as gpd
import os
import pandas as pd
```

```python
df3['읍면동'] = df3['읍면동'].apply(lambda x: re.sub(r'[·.]', '', x))
df3[df3['읍면동'].str.contains('[·.]')]
```

```python
dong_shp = gpd.read_file('/content/drive/MyDrive/new_project1/data/seoul_shp/dong/bnd_dong_11_2023_2023_2Q.shp')
dong_shp.head()
```

```python
df3['구동'] = df3['시군구'] + ' ' +df3['읍면동']
df3.head()
```

```python
df3['대학교 개수'] = df3['대학교 개수'].apply(lambda x: 0 if x =='-' else x)
df3['대학교 개수'] = df3['대학교 개수'].apply(lambda x: int(x))
df3['대학교 개수'].unique()
```

```python
df3 = df3[['구동', '대학교 개수']]
df3
```

```python
region_code = pd.read_csv('/content/drive/MyDrive/new_project1/data/complete/region_code.csv')
region_code.head()
```

```python
dong_shp['ADM_CD'] = dong_shp['ADM_CD'].astype('int')
```

```python
region_code['ADM_CD'] = region_code['ADM_CD'].astype('int')
```

```python
map_base = dong_shp.merge(region_code,
                          how = 'left',
                          on = 'ADM_CD')
map_base.head()
```

```python
map_base['ADM_NM'] = map_base['ADM_NM'].apply(lambda x: re.sub(r'\·', '', x))
```

```python
map_base['key1'] = map_base['시군구'] + ' ' + map_base['ADM_NM']
map_base['key2'] = map_base['시군구'] + ' ' + map_base['읍면동']
```

```python
map_base[['key1', 'key2']]
```

```python
final4 = map_base.merge(df3.rename(columns = {'구동' : 'key1'}),
                       how = 'left',
                       on = 'key1')
final4
```
