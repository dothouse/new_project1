```python
year_list = ['2022', '2023', '2024']


for year in year_list:

    solo_col = solo.columns[solo.columns.str.contains(year)].tolist()
    solo_col = ['ADM_CD'] + solo_col
    temp_solo = solo[solo_col[:11]]
    temp_solo[f'solo_20s_{year}'] = temp_solo.apply(lambda x: (x[f'solo_20_{year}'] + x[f'solo_25_{year}']) / 2, axis = 1)
    temp_solo[f'solo_30s_{year}'] = temp_solo.apply(lambda x: (x[f'solo_30_{year}'] + x[f'solo_35_{year}']) / 2, axis = 1)

    temp_solo[f'solo_all_ratio'] = temp_solo.apply(lambda x: x[f'solo_all_{year}'] / x[f'total_all_{year}'] * 100, axis=1)
    temp_solo[f'solo_20s_ratio'] = temp_solo.apply(lambda x: x[f'solo_20s_{year}'] / x[f'solo_all_{year}'] * 100, axis=1)
    temp_solo[f'solo_30s_ratio'] = temp_solo.apply(lambda x: x[f'solo_30s_{year}'] / x[f'solo_all_{year}'] * 100, axis=1)

    globals()[f'solo_{year}'] = temp_solo.copy()
    globals()[f'solo_{year}'].to_csv(f'{data_path}/year/solo_{year}.csv', index = False)
```

```python
store = pd.read_csv(f'{data_path}/final/store.csv')
store.head()
```

```python
for year in year_list:

    store_col = store.columns[store.columns.str.contains(year)].tolist()
    store_col = ['ADM_CD'] + store_col
    temp_store = store[store_col]

    temp_store[f'네일숍_{year}'] =  temp_store[f'미용업 네일아트업_{year}']
    temp_store[f'미용실_{year}'] =  temp_store[f'미용업 일반미용업_{year}']
    temp_store[f'분식전문점_{year}'] =  temp_store[f'일반음식점 김밥(도시락)_{year}'] + temp_store[f'일반음식점 분식_{year}']
    temp_store[f'애완동물_{year}'] =  temp_store[f'동물미용업_{year}'] +temp_store[f'동물병원_{year}'] + temp_store[f'동물약국_{year}']
    temp_store[f'양식음식점_{year}'] =  temp_store[f'일반음식점 경양식_{year}'] + temp_store[f'일반음식점 외국음식전문점(인도,태국등)_{year}']
    temp_store[f'일식음식점_{year}'] =  temp_store[f'일반음식점 일식_{year}']
    temp_store[f'제과점_{year}'] =  temp_store[f'제과점영업 제과점영업_{year}']
    temp_store[f'중식음식점_{year}'] =  temp_store[f'일반음식점 중국식_{year}']
    temp_store[f'치킨전문점_{year}'] =  temp_store[f'일반음식점 통닭(치킨)_{year}']
    temp_store[f'커피-음료_{year}'] =  temp_store[f'일반음식점 까페_{year}'] + temp_store[f'일반음식점 커피숍_{year}']
    temp_store[f'패스트푸드점_{year}'] =  temp_store[f'일반음식점 패스트푸드_{year}'] +temp_store[f'휴게음식점 패스트푸드_{year}']
    temp_store[f'편의점_{year}'] =  temp_store[f'휴게음식점 편의점_{year}']
    temp_store[f'한식음식점_{year}'] =  temp_store[f'일반음식점 한식_{year}']
    temp_store[f'호프-간이주점_{year}'] =  temp_store[f'일반음식점 감성주점_{year}']  + temp_store[f'일반음식점 정종/대포집/소주방_{year}']


    store_newvar = [f'네일숍_{year}', f'미용실_{year}', f'분식전문점_{year}', f'애완동물_{year}',
                    f'한식음식점_{year}', f'양식음식점_{year}', f'일식음식점_{year}',  f'중식음식점_{year}', f'치킨전문점_{year}',
                    f'커피-음료_{year}', f'패스트푸드점_{year}', f'제과점_{year}', f'편의점_{year}',  f'호프-간이주점_{year}']

    temp_store_col = ['ADM_CD'] + store_newvar
    globals()[f'store_{year}'] = temp_store[temp_store_col]
    globals()[f'store_{year}'].to_csv(f'{data_path}/year/store_{year}.csv', index = False)

```

# 연도별로 결합

```python
for year in year_list:
    solo = pd.read_csv(f'{data_path}/year/solo_{year}.csv')
    store = pd.read_csv(f'{data_path}/year/store_{year}.csv')
    real = pd.read_csv(f'{data_path}/year/real_{year}.csv')
    seoul = pd.read_csv(f'{data_path}/year/seoul_{year}.csv')
    sang = pd.read_csv(f'{data_path}/year/sang_{year}.csv')

    df_list = ['solo', 'store', 'real', 'seoul', 'sang']

    merged_df = pd.DataFrame()

    for idx, df_name in enumerate(df_list):
        df = globals()[df_name]
        if idx == 0:
            merged_df = df
        else :
            merged_df = merged_df.merge(df,
                                        how = 'left',
                                        on = 'ADM_CD')

    globals()[f'final_{year}'] = merged_df.copy()
    globals()[f'final_{year}'].to_csv(f'{data_path}/final/final_{year}.csv', index = False)

```
