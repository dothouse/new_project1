```python
# 2022, 2023년 선택
income = income[(income['기준_년분기_코드'] == 20231) | (income['기준_년분기_코드'] == 20221)]
income.head()
```

```python
income_year_list = ['2022', '2023']

for year in income_year_list:
    income_var = income.columns[5:].tolist()
    income_var_new = []
    for var in income_var:
        temp_var = var.split('_')[0] + f'_{year}'
        income_var_new.append(temp_var)
    globals()[f'income_{year}'].columns = ['기준년도', '행정동코드', '행정동', f'월평균소득_{year}', f'소득구간_{year}'] + income_var_new
    globals()[f'income_{year}'].drop(columns = '기준년도', inplace = True)
```

```python
income = pd.merge(income_2022, income_2023,
                  how = 'left',
                  on = ['행정동코드', '행정동'])
income.drop(columns = '행정동', inplace = True)
income
```
